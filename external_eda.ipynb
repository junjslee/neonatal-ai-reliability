```python
# %% 라이브러리 임포트
import pandas as pd
import matplotlib.pyplot as plt

# %% 데이터 불러오기
# CSV 파일 경로 (네트워크 경로)
file_path = r"\\192.168.45.125\forGPU2\yeonsu\0.Projects\Pneumoperitoneum\External EDA\External.csv"
df = pd.read_csv(file_path)

```

```python
# %% width와 height의 최소값, 최대값 출력
width_min = df['Width'].min()
width_max = df['Width'].max()
height_min = df['Height'].min()
height_max = df['Height'].max()

print("Width - min:", width_min, ", max:", width_max)
print("Height - min:", height_min, ", max:", height_max)

```

```python
# %% H 컬럼의 값에 따른 count 및 percentage 계산
counts = df['H'].value_counts().sort_index()  # H 컬럼의 각 값에 대한 빈도수 계산
total = counts.sum()  # 전체 개수 (또는 len(df))
percentages = counts / total * 100

# %% 결과 출력
print("H 컬럼 값별 통계:")
for value, count in counts.items():
    percentage = count / total * 100
    print(f"{value}: count = {count}, percentage = {percentage:.2f}%")

# %% 막대그래프 생성 및 수치 표시
plt.figure(figsize=(8, 6))
bars = plt.bar(counts.index.astype(str), counts.values, color='skyblue')
plt.xlabel("H Distribution")
plt.ylabel("Count")
plt.title("H Distribution")

# 각 막대 위에 count와 percentage 표시
for bar, count in zip(bars, counts):
    percentage = count / total * 100
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(), 
             f"{count}\n({percentage:.1f}%)", 
             ha='center', va='bottom', fontsize=10)

plt.show()
```
